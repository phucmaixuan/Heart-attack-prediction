```python
import matplotlib.pyplot as plt
import numpy as np

# 1) Điền accuracy của từng mô hình ở đây
models = ["SVM", "Naive Bayes", "Random Forest", "KNN", "ANN"]
precision = [
    0.6521,  # SVM
    0.6527,  # Naive Bayes
    0.6561,  # Random Forest
    0.1702,  # KNN
    0.6524   # ANN
]
# ↑ THAY các số phía trên bằng accuracy thật của nhóm bạn

# 2) Vẽ biểu đồ
x = np.arange(len(models))

plt.figure(figsize=(8,5))
bars = plt.bar(x, precision, color='orange')

# Ghi số lên trên từng cột
for i, v in enumerate(precision):
    plt.text(
        x[i],               # vị trí theo trục x
        v + 0.005,          # cao hơn đỉnh cột một chút
        f"{v:.4f}",         # format 4 chữ số thập phân
        ha='center',
        va='bottom',
        fontsize=10
    )

plt.xticks(x, models, rotation=15)
plt.ylim(0, 1.05)
plt.ylabel("Precision")
plt.title("So sánh Precision giữa các mô hình")

# Thêm lưới ngang
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```
