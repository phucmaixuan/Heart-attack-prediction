```python
import matplotlib.pyplot as plt
import numpy as np

# 1) Điền recall của từng mô hình ở đây
models = ["SVM", "Naive Bayes", "Random Forest", "KNN", "ANN"]
recall = [
    0.9888,  # SVM
    0.9990,  # Naive Bayes
    0.8924,  # Random Forest
    0.0327,  # KNN
    0.9932   # ANN
]
# ↑ THAY các số phía trên bằng recall thật của nhóm bạn

# 2) Vẽ biểu đồ
x = np.arange(len(models))

plt.figure(figsize=(8,5))
bars = plt.bar(x, recall, color='green')

# Ghi số lên trên từng cột
for i, v in enumerate(recall):
    plt.text(
        x[i],               # vị trí theo trục x
        v + 0.005,          # cao hơn đỉnh cột một chút
        f"{v:.4f}",         # format 4 chữ số thập phân
        ha='center',
        va='bottom',
        fontsize=10
    )

plt.xticks(x, models, rotation=15)
plt.ylim(0, 1.05)
plt.ylabel("Recall")
plt.title("So sánh Recall giữa các mô hình")

# Thêm lưới ngang
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```
