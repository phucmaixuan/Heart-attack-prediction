```python
import matplotlib.pyplot as plt
import numpy as np

# 1) Điền recall của từng mô hình ở đây
models = ["SVM", "Naive Bayes", "Random Forest", "KNN", "ANN"]
F1_score = [
    0.7859,  # SVM
    0.7895,  # Naive Bayes
    0.7562,  # Random Forest
    0.0548,  # KNN
    0.7875   # ANN
]
# ↑ THAY các số phía trên bằng recall thật của nhóm bạn

# 2) Vẽ biểu đồ
x = np.arange(len(models))

plt.figure(figsize=(8,5))
bars = plt.bar(x, F1_score, color='pink')
# Ghi số lên trên từng cột
for i, v in enumerate(F1_score):
    plt.text(
        x[i],               # vị trí theo trục x
        v + 0.005,          # cao hơn đỉnh cột một chút
        f"{v:.4f}",         # format 4 chữ số thập phân
        ha='center',
        va='bottom',
        fontsize=10
    )

plt.xticks(x, models, rotation=15)
plt.ylim(0, 1.05)
plt.ylabel("F1-Score")
plt.title("So sánh F1-Score giữa các mô hình")

# Thêm lưới ngang
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

models = ["SVM", "Naive Bayes", "Random Forest", "KNN", "ANN"]

# Số liệu của NHÓM BẠN
accuracy = [0.9705, 0.9708, 0.9690, 0.9383, 0.9707]
precision = [0.6521, 0.6527, 0.6561, 0.1702, 0.6524]
recall    = [0.9888, 0.9990, 0.8924, 0.0327, 0.9932]
f1        = [0.7859, 0.7895, 0.7562, 0.0548, 0.7875]

x = np.arange(len(models))

fig, axes = plt.subplots(2, 2, figsize=(10, 8))
fig.suptitle("Model Performance Comparison", fontsize=14)

# Hàm tiện để vẽ bar + ghi số + lưới
def plot_bar(ax, values, title, color):
    ax.bar(x, values, color=color)
    ax.set_title(title)
    ax.set_ylim(0, 1.05)
    ax.set_xticks(x)
    ax.set_xticklabels(models, rotation=15)
    ax.grid(axis='y', linestyle='--', alpha=0.5)
    # Ghi số lên cột
    for i, v in enumerate(values):
        ax.text(
            x[i],
            v + 0.01,
            f"{v:.4f}",
            ha='center',
            va='bottom',
            fontsize=8
        )

# 1. Accuracy (xanh dương)
plot_bar(axes[0, 0], accuracy, "Accuracy", 'tab:blue')

# 2. Precision (cam)
plot_bar(axes[0, 1], precision, "Precision", 'orange')

# 3. Recall (xanh lá)
plot_bar(axes[1, 0], recall, "Recall", 'green')

# 4. F1-score (hồng nhạt)
plot_bar(axes[1, 1], f1, "F1-Score", '#FFB6C1')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

# Nếu muốn lưu 1 hình duy nhất để đưa vào LaTeX:
# fig.savefig("Picture/Evaluation/model_performance_comparison.png", dpi=300)
```
